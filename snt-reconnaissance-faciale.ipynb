-->Pour l’instant, l’algorithme a créé une image de dégradé de gris (1ère partie ci-dessus).

    Comment transforme t’il cette image 1ère partie en cette autre image 2ème partie ?

==>Comparez les zones A et B dans la 1ère puis la 2ème partie et proposez une explication :

Si on analyse une nouvelle fois verticalement, de haut en bas, cette image 1ère partie, on voit que la partie A est plutôt calme : elle est constituée d’un dégradé très doux d’un gris moyen vers le gris clair : sans aucun ‘accident’ de contraste. Par contre, en zone B, on passe assurément brusquement d’une zone gris clair à une zone gris foncé ; dans la 2ème partie, cela se concrétise par l’absence d’apparition d’une ligne noire en zone A, et l’apparition d’une ligne noire en zone B : on peut donc dire que la 2ème partie est le résultat d’un algorithme de contraste/contour appliqué à la 1ère partie.
Ainsi, il n’y a pas de ligne horizontale caractéristique observée au front de la jeune femme, mais il y en a une située à la hauteur de ses narines.
